```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
 


my_url= "https://www.newegg.com/Product/ProductList.aspx?Submit=ENE&N=100007709%2050001312%2050001314%2050001315%2050001402%2050001419%2050001471%2050001561%2050001944%2050012150%204814%20601201888%20601204369%20601301599%20601296379%20601296377%2050001669%20601321570%20601321572%20601323902%20601328427%20601331000%20601331379&IsNodeId=1&cm_sp=Cat_video-Cards_1-_-Visnav-_-Gaming-Video-Cards_2"

#Opens up a connection
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()


# html parsing
page_soup = soup(page_html, "html.parser")


```

```python
# We can see the several differnt html tags here
#Examples of tags are h1,h3,p, body.span etc. 
page_soup.p
```

```python
#Grabs each product
containers = page_soup.findAll("div",{"class":"item-container"})
```

```python
#can check length of containers
#len(containers)
container = containers[0]

divWithInfo = containers[0].find("div","item-info")

#iterative for loop

for container in containers: 
    brand = divWithInfo.div.a.img["title"]
    
    title_container = container.findAll("a", {"class":"item-title"})
    product_name = title_container[0].text
    
   # shipping_container = container.findAll("li", {"class" :"price-ship" })
    shipping_container = container.findAll("li", {"class" :"price-ship" })
    shipping_container[0].text.strip()
    shipping_container[0].text.strip("[<li class="price-ship'>")
    
    
   # print("Brand: " + str(brand))
   # print("Product name: " + str(product_name))
    print("Shipping option: " + str(shipping_container))
```
